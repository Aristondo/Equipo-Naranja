```python
merged_df = pd.merge(df_2022, df_2021, on='Código NIF', how='inner')
merged_df = pd.merge(merged_df, df_2020, on='Código NIF', how='inner')

# Get businesses present in all three dataframes
common_businesses = merged_df['Código NIF']

# Concatenate dataframes to find all businesses
all_businesses = pd.concat([df_2022['Código NIF'], df_2021['Código NIF'], df_2020['Código NIF']])

# Find businesses that are not common
unique_businesses = all_businesses[~all_businesses.isin(common_businesses)]

# Identify which dataset each unique business belongs to
df_2022_unique = df_2022[df_2022['Código NIF'].isin(unique_businesses)]
df_2021_unique = df_2021[df_2021['Código NIF'].isin(unique_businesses)]
df_2020_unique = df_2020[df_2020['Código NIF'].isin(unique_businesses)]
```

```python
# Identify businesses present in only two dataframes
common_df_2022_2021 = pd.merge(df_2022, df_2021, on='Código NIF', how='inner')
common_df_2022_2020 = pd.merge(df_2022, df_2020, on='Código NIF', how='inner')
common_df_2021_2020 = pd.merge(df_2021, df_2020, on='Código NIF', how='inner')

# Combine businesses present in only two dataframes
common_in_two = pd.concat([common_df_2022_2021['Código NIF'], common_df_2022_2020['Código NIF'], common_df_2021_2020['Código NIF']])

# Find unique businesses in each pair
df_2022_2021_unique = common_df_2022_2021[~common_df_2022_2021['Código NIF'].isin(common_in_two)]
df_2022_2020_unique = common_df_2022_2020[~common_df_2022_2020['Código NIF'].isin(common_in_two)]
df_2021_2020_unique = common_df_2021_2020[~common_df_2021_2020['Código NIF'].isin(common_in_two)]
```
